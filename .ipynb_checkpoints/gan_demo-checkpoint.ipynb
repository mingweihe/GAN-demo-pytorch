# Generate a curve artwork by using GAN

```python
# import sys
# !python --version
# !python3 --version
# !{sys.executable} --version
# !{sys.executable} -m pip install torch
# !{sys.executable} -m pip install torchsummary
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from torchsummary import summary
%matplotlib inline
```

```python
# hyper parameters
BATCH_SIZE = 64
LR_G = .0001
LR_D = .0001
N_IDEAS = 5 # input size
ART_COMPONENTS = 15 # output size
PAINT_POINTS = np.vstack([np.linspace(-1, 1, ART_COMPONENTS) for _ in range(BATCH_SIZE)])
TOTAL_STEPS = 2000
SHOW_INTERVAL = TOTAL_STEPS / 10
```

```python
def artist_works():
    r = .05 * np.random.randn(1, ART_COMPONENTS)
    paintings = np.sin(PAINT_POINTS * np.pi) + r
    paintings = torch.from_numpy(paintings).float()
    return paintings
```

```python
# show one of the paintings
plt.plot(PAINT_POINTS[0], artist_works()[0])
plt.show()
```

```python
G = nn.Sequential(
    nn.Linear(N_IDEAS, 128),
    nn.ReLU(),
    nn.Linear(128, ART_COMPONENTS),
)

D = nn.Sequential(
    nn.Linear(ART_COMPONENTS, 128),
    nn.ReLU(),
    nn.Linear(128, 1),
    nn.Sigmoid(),
)
```

```python
# summary(G, (1, N_IDEAS))
```

```python
# summary(D, (1, ART_COMPONENTS))
```

```python
opt_D = torch.optim.Adam(D.parameters(), lr=LR_D)
opt_G = torch.optim.Adam(G.parameters(), lr=LR_G)
plt.ion()
D_loss_history = []
G_loss_history = []
for step in range(TOTAL_STEPS):
    artist_paintings = artist_works()
    G_ideas = torch.randn(BATCH_SIZE, N_IDEAS)
    G_paintings = G(G_ideas)
    prob_artist0 = D(artist_paintings)
    prob_artist1 = D(G_paintings)
    D_loss = -torch.mean(torch.log(prob_artist0) + torch.log(1. -prob_artist1))
    G_loss = torch.mean(torch.log(1. -prob_artist1))
    D_loss_history.append(D_loss)
    G_loss_history.append(G_loss)
    opt_D.zero_grad() 
    D_loss.backward(retain_graph=True)
    opt_D.step() # update the network parameters
    opt_G.zero_grad()
    G_loss.backward()
    opt_G.step()
    # plotting
    if step % SHOW_INTERVAL == 0:
        plt.cla()
        plt.plot(PAINT_POINTS[0], G_paintings.data.numpy()[0], label='Generated paiting')
        plt.plot(PAINT_POINTS[0], artist_paintings[0], label='Artist\'s paiting')
        plt.text(-1, .75, 'D accuracy=%.2f (0.5 for D to converage)'
                % prob_artist0.data.numpy().mean(), fontdict={'size': 8})
        plt.text(-1, .5, 'D loss=%.2f (1.38 for G to converage)'
                % D_loss.data.numpy(), fontdict={'size': 8})
        plt.ylim((-1, 1))
        plt.legend(loc='lower right')
plt.ioff()
plt.show()
```

```python
# discriminator loss
plt.plot(D_loss_history)
plt.show()
```

```python
# generator loss
plt.plot(G_loss_history)
plt.show()
```

```python
# test generator
curve = G(torch.randn(N_IDEAS))
plt.plot(curve.data.numpy())
```
